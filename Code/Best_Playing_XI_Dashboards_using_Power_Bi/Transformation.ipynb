```python
import pandas as pd
import json
```

```python
# Batting Score Path
json_path="json/match_summary.json"

with open(json_path) as f:
    data = json.load(f)

df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
# Dimension
df_match.shape
```

```python
# Rename column
df_match.rename({'match_id':'scorecard'}, axis = 1,inplace=True)
df_match.head()
df_match.rename({'scorecard':'match_id'},axis = 1,inplace =True)
df_match.head()
```

```python
# match_id (combinning two tables)

match_ids_dict = {}

for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

print(match_ids_dict)
```

### batting_summary

```python
batting_path="json/batting_score.json"

with open(batting_path) as f:
    data = json.load(f)

    all_records = []

    for rec in data:
        all_records.extend(rec['battingSummary'])
df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
# Rename Column
df_batting.rename({'out/not_out':'dismissal'},axis=1,inplace=True)
df_batting.head()
```

```python
# Create out/not_out column based on dismissal
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head()
```

```python
# Remove Column
df_batting.drop(columns={"dismissal"}, inplace = True)
df_batting.head(100)
```

```python
# Replace any word
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('ae',''))
df_batting.head()
```

```python
df_batting.drop(columns="match_id",inplace=True)
```

```python
df_batting.head()
```

```python
# adding match_id column into df_batting by match_ids_dict
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head()
```
